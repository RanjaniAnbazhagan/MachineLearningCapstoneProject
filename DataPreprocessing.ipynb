```python
# Step 1: Import libraries
import pandas as pd
import numpy as np

# Step 2: Load datasets
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

# Step 3: Inspect data
print(train.head())
print(train.info())
print(train.describe())
```

```python
#Check missing values
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
# Fill missing Age with median
train['Age']=train['Age'].fillna(train['Age'].median())
test['Age']=test['Age'].fillna(test['Age'].median())
```

```python
# Fill missing Embarked with mode
train['Embarked']=train['Embarked'].fillna(train['Embarked'].mode()[0])

# Fill missing Fare in test with median
test['Fare']=test['Fare'].fillna(test['Fare'].median())

```

```python
# Step 4: Check missing values
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
# Drop Cabin (too many missing values)
train=train.drop(columns=['Cabin','Name','Ticket'])
test=test.drop(columns=['Cabin','Name','Ticket'])
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

```python
dataset_train= pd.get_dummies(train, columns=['Sex','Embarked'], dtype=int, drop_first=True)
```

```python
dataset_train
```

```python
dataset_test= pd.get_dummies(test, columns=['Sex','Embarked'], dtype=int, drop_first=True)
```

```python
dataset_test
```

```python
#Input & Output coloumn split
independent=dataset_train[['PassengerId', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Sex_male', 'Embarked_Q', 'Embarked_S']]
dependent=dataset_train['Survived']
```

```python
from sklearn.model_selection import train_test_split
#calling the parameterized function train_test_split 
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
Classifier=RandomForestClassifier(n_estimators=10,criterion='entropy',max_features=None)
Classifier.fit(independent,dependent)
```

```python
y_pred=Classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report=classification_report(y_test,y_pred)
```
