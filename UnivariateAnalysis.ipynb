```python
import pandas as pd
import numpy as np

# Step 2: Load datasets
Dataset = pd.read_csv("train_preprocessed.csv")
```

```python
Dataset
```

```python
dupdataset=pd.DataFrame()
```

```python
dupdataset.loc['Mean', 'Age']   = Dataset['Age'].mean()
dupdataset.loc['Median', 'Age'] = Dataset['Age'].median()
dupdataset.loc['Mode', 'Age']   = Dataset['Age'].mode()[0]

print(dupdataset)
```

```python
import numpy as np
np.percentile(Dataset["Age"],100)
```

```python
dupdata=pd.DataFrame(index=["mean","median","mode","Q1:25%","Q2:50%","Q3:75%","99","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"],columns=['Age','Fare'])
for columnName in ['Age','Fare']:
        dupdata.loc["mean", columnName]   = Dataset[columnName].mean()
        dupdata.loc["median", columnName] = Dataset[columnName].median()
        dupdata.loc["mode", columnName]   = Dataset[columnName].mode()[0]
        dupdata.loc["Q1:25%", columnName] = Dataset[columnName].quantile(0.25)
        dupdata.loc["Q2:50%", columnName] = Dataset[columnName].quantile(0.50)
        dupdata.loc["Q3:75%", columnName] = Dataset[columnName].quantile(0.75)
        dupdata.loc["99", columnName]     = np.percentile(Dataset[columnName], 99)
        dupdata.loc["Q4:100%", columnName]= Dataset[columnName].max()
        dupdata.loc["IQR", columnName]    = dupdata.loc["Q3:75%", columnName] - dupdata.loc["Q1:25%", columnName]
        dupdata.loc["1.5rule", columnName]= 1.5 * dupdata.loc["IQR", columnName]
        dupdata.loc["Lesser", columnName] = dupdata.loc["Q1:25%", columnName] - dupdata.loc["1.5rule", columnName]
        dupdata.loc["Greater", columnName]= dupdata.loc["Q3:75%", columnName] + dupdata.loc["1.5rule", columnName]
        dupdata.loc["Min", columnName]    = Dataset[columnName].min()
        dupdata.loc["Max", columnName]    = Dataset[columnName].max()
```

```python
lesser = []
greater = []

for columnName in ['Age','Fare']:
    if Dataset[columnName].min() < dupdata.loc["Lesser", columnName]:
        lesser.append(columnName)
    if Dataset[columnName].max() > dupdata.loc["Greater", columnName]:
        greater.append(columnName)
```

```python
Dataset_original = Dataset.copy()
```

```python
for columnName in lesser:
    Dataset.loc[Dataset[columnName] < dupdata.loc["Lesser", columnName], columnName] = dupdata.loc["Lesser", columnName]

for columnName in greater:
    Dataset.loc[Dataset[columnName] > dupdata.loc["Greater", columnName], columnName] = dupdata.loc["Greater", columnName]
```

```python
dupdata.loc["Min", columnName] = Dataset_original[columnName].min()
dupdata.loc["Max", columnName] = Dataset_original[columnName].max()
```
